1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:



```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return x**2 - 4*x + 2

def f2(x):
    return np.sin(3*x)

def f3(x):
    return np.exp(2*x) - 3

def f4(x):
    return np.sqrt(x**2 - 4)

def f5(x):
    return np.cos(x + np.pi/4)

x = np.linspace(-3, 3, 101)
x_pos = np.linspace(2, 10, 100)
```

```python
plt.figure(figsize=(3, 2))
plt.plot(x, f1(x), label=r'$f(x) = x^2 - 4x + 2$')
plt.title("f(x) = x^2 - 4x + 2")
plt.grid()
```

Dominio = Todos los números reales

```python
plt.figure(figsize=(3, 2))
plt.plot(x, f2(x), label=r'$f(x) = \sin(3x)$')
plt.title("f(x) = sin(3x)")
plt.grid()
```

Dominio = Todos los números reales

```python
plt.figure(figsize=(3, 2))
plt.plot(x, f3(x), label=r'$f(x) = e^{2x} - 3$')
plt.title("f(x) = e^{2x} - 3")
plt.grid()
```

Dominio = Todos los números reales

```python
plt.figure(figsize=(3, 2))
plt.plot(x_pos, f4(x_pos), label=r'$f(x) = \sqrt{x^2 - 4}$')
plt.title("f(x) = √(x^2 - 4)")
plt.grid()
```

```python
plt.figure(figsize=(3, 2))
plt.plot(x, f5(x), label=r'$f(x) = \cos(x + \frac{\pi}{4})$')
plt.title("f(x) = cos(x + π/4)")
plt.grid()
```

Dominio = Todos los números reales

2. Graficar los siguientes grupos de funciones:

```python
x = np.linspace(-3, 3, 101)

y1 = 2*x + 1
y2 = -x + 4
y3 = x**2

plt.figure(figsize=(5, 3))

plt.plot(x, y1, label='y1 = 2x + 1', color='blue')
plt.plot(x, y2, label='y2 = -x + 4', color='red')
plt.plot(x, y3, label='y3 = x^2', color='green')

plt.grid()

plt.xlabel('x')
plt.ylabel('y')
plt.title('Grupo 1')

plt.legend()
```

```python
x = np.linspace(-3, 3, 101)

y1_g2 = np.sin(x)
y2_g2 = np.cos(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y1_g2, label=r'$y_1 = \sin(x)$')
plt.plot(x, y2_g2, label=r'$y_2 = \cos(x)$')
plt.title('Grupo 2')
plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend()
```

3. Crea una función que cree una gráfica y dibuje 'k' líneas de forma a. La función recibe:

x1: valor inicial de 'x'

x2: valor final de 'x'

n: cantidad de puntos del vector 'x'

a: valor inicial de a

b: valor final de a

k: cantidad de líneas

Usar el la variable opcional de "label" y el comando de legend para saber qué línea pertenece a qué.

```python
def foo(x1, x2, n, a_inicio, a_fin, k):
    x = np.linspace(x1, x2, n)
    alphas = np.linspace(a_inicio, a_fin, k)
    
    for alpha in alphas:
        y = np.sin(alpha * x)
        plt.plot(x, y, label=f'α = {alpha:.1f}')
    
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()

foo(0, 2 * np.pi, 100, 1, 7, 7)
```
